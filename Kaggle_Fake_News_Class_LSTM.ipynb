```python
## import the lib : stemming, stopwords
## we ar not doing lemmatization as it takes time as the data is too big

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
ps1 = PorterStemmer()

corpus = []
```

```python
## feature eng

for i in range(0, len(df1)):
  rev1 = re.sub('[^a-zA-z]', ' ', df1['title'][i])
  rev1 = rev1.lower()
  rev1 = rev1.split()

  rev1 = [ps1.stem(word) for word in rev1 if not word in stopwords.words('english')]
  rev1 = ' '.join(rev1)
  corpus.append(rev1)
```
