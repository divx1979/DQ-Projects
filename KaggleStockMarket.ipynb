```python
import numpy as np
import pandas as pd
```

```python
## Data taken from Kaggle Daily News And Stock Pred
# https://www.kaggle.com/datasets/aaron7sun/stocknews
```

```python
data = pd.read_csv('/content/stock_nlp.csv')
```

```python
data.head()
```

```python
# Now let's divide the data into train and test

train = data[data['Date'] < '20150101']
test = data[data['Date'] > '20141231']
```

```python
# Now, let's preprocess data by selecting the cols and cleaning up the punctuations
## and then renaming the index using nums

data1 = train.iloc[:, 2:27]
```

```python
data1.replace('[^a-zA-Z]', ' ', regex = True, inplace = True)
```

```python
new_list = [ i for i in range(25)]
new_index = [ str(i) for i in new_list]
data1.columns = new_index
```

```python
data1.head()
```

```python
for Index in new_index:
  data1[Index] = data1[Index].str.lower()

data1
```

```python
headline = []
for row in range(0, len(data1.index)):
  headline.append(' '.join(str(i) for i in data1.iloc[row, 0:25]))
```

```python
headline[1]
```

```python
## Now, let's apply the same techniques to the test data

data2 = test.iloc[:, 2:27]
```

```python
data2.replace('[^a-zA-Z]', ' ', regex = True, inplace = True)
```

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer(ngram_range = (2,2))
traindata = cv.fit_transform(headline)
```

```python
rfc = RandomForestClassifier(n_estimators = 200, criterion = 'entropy')
rfc.fit(traindata, train['Label'])
```

```python
testdata = cv.transform(test_transform)
pred = rfc.predict(testdata)
```

```python
#checking scores of accuracy

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# check confusion mat
cm = confusion_matrix(test['Label'], pred)

# check accuracy sco
ac = accuracy_score(test['Label'], pred)

# check classification
cl = classification_report(test['Label'], pred)
```
